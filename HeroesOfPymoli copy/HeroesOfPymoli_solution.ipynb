```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load
file_to_load = os.path.join("Resources", "purchase_data.csv")

# Read Purchasing File and store into Pandas data frame
gamingDF = pd.read_csv(file_to_load)
gamingDF.head()
```

```python
uniqueSNs = gamingDF["SN"].unique()
numUniqueSNs = uniqueSNs.size
playerCount = pd.DataFrame({"Player Count": [numUniqueSNs]})

playerCount.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueItems = gamingDF["Item ID"].unique()

avgPrice = gamingDF["Price"].mean()

numPurchases = gamingDF["Purchase ID"].unique()

totRevenue = gamingDF["Price"].sum()

purchasingAnalysis = pd.DataFrame({
    "Number of Unique Items": [uniqueItems.size],
    "Average Price": [avgPrice],
    "Number of Purchases": [numPurchases.size],
    "Total Revenue": [totRevenue]
})

purchasingAnalysis["Average Price"] = purchasingAnalysis["Average Price"].astype(float).map("${:,.2f}".format)
purchasingAnalysis["Total Revenue"] = purchasingAnalysis["Total Revenue"].astype(float).map("${:,.2f}".format)

purchasingAnalysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# gets the genders of every unique SN, shrinks DF so there are no repeat SNs
uniqueNameDF = gamingDF.drop_duplicates(subset='SN', keep='first')

# collects and sums the # of each gender within reduced DF
females = uniqueNameDF.loc[uniqueNameDF["Gender"] == "Female", :]
males = uniqueNameDF.loc[uniqueNameDF["Gender"] == "Male", :]
oth = uniqueNameDF.loc[uniqueNameDF["Gender"] == "Other / Non-Disclosed", :]

numFemale = len(females)
numMale = len(males)
numOth = len(oth)

# coumpute % of each gender
perFemale = (numFemale/numUniqueSNs)*100
perMale = (numMale/numUniqueSNs)*100
perOth = (numOth/numUniqueSNs)*100

# toss gender counts, percentages in new DF with percentage formatting
genderDemog = pd.DataFrame({
    "Total Count" : [numFemale, numMale, numOth],
    "Percentage of Players" : [perFemale, perMale, perOth]
})

# set corresponding gender for each count as only the index, rather than its own column
genderDemog["Percentage of Players"] = genderDemog["Percentage of Players"].astype(float).map("{:,.2f}%".format)
genderIndexedDF = genderDemog.set_index([pd.Index(["Female", "Male", "Other / Non-Disclosed"])])

genderIndexedDF.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
femaleSet = gamingDF.loc[gamingDF["Gender"] == "Female", :]
maleSet = gamingDF.loc[gamingDF["Gender"] == "Male", :]
othSet = gamingDF.loc[gamingDF["Gender"] == "Other / Non-Disclosed", :]

femPurch = len(femaleSet)
malPurch = len(maleSet)
othPurch = len(othSet)

avgFemSpent = femaleSet["Price"].mean()
avgMaleSpent = maleSet["Price"].mean()
avgOthSpent = othSet["Price"].mean()

totFemSpent = femaleSet["Price"].sum()
totMaleSpent = maleSet["Price"].sum()
totOthSpent = othSet["Price"].sum()

peruserFemSpent = totFemSpent/numFemale
peruserMaleSpent = totMaleSpent/numMale
peruserOthSpent = totOthSpent/numOth

genderPurchDemog = pd.DataFrame({
    "Purchase Count" : [femPurch, malPurch, othPurch],
    "Average Purchase Price" : [avgFemSpent, avgMaleSpent, avgOthSpent],
    "Total Purchase Value" : [totFemSpent, totMaleSpent, totOthSpent],
    "Average Total Purchase per Person" : [peruserFemSpent, peruserMaleSpent, peruserOthSpent]
})


genderPurchDemog["Average Purchase Price"] = genderPurchDemog["Average Purchase Price"].astype(float).map("${:,.2f}".format)
genderPurchDemog["Total Purchase Value"] = genderPurchDemog["Total Purchase Value"].astype(float).map("${:,.2f}".format)
genderPurchDemog["Average Total Purchase per Person"] = genderPurchDemog["Average Total Purchase per Person"].astype(float).map("${:,.2f}".format)

genderIndexedDF = genderPurchDemog.set_index([pd.Index(["Female", "Male", "Other / Non-Disclosed"])])

genderIndexedDF.head()

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


uniqueNameDF["Age Ranges"] = pd.cut(gamingDF["Age"].astype(int), bins, labels=group_names)
#binnedAgesDF = uniqueNameDF["Age Ranges"]

ageRangeData = uniqueNameDF.groupby("Age Ranges").count()
ageRangeData["Percentage of Players"] = (ageRangeData["SN"]/numUniqueSNs)*100

ageRangeData = ageRangeData.rename(columns={"SN" : "Player Count"})

ageRangeData = ageRangeData[["Player Count", "Percentage of Players"]]

ageRangeData["Percentage of Players"] = ageRangeData["Percentage of Players"].astype(float).map("{:,.2f}%".format)

ageRangeData
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gamingDF["Age Ranges"] = pd.cut(gamingDF["Age"].astype(int), bins, labels=group_names)

groupedGamingDF = pd.DataFrame(gamingDF.groupby("Age Ranges")["SN"].count())
groupedGamingDF = groupedGamingDF.rename(columns={"SN" : "Purchase Count"})
groupedGamingDF["Total Purchase Value"] = gamingDF.groupby("Age Ranges")["Price"].sum()
groupedGamingDF["Average Purchase Price"] = gamingDF.groupby("Age Ranges")["Price"].mean()
groupedGamingDF["Avg Total Purchase per Person"] = groupedGamingDF["Total Purchase Value"] / (ageRangeData["Player Count"])

groupedGamingDF["Average Purchase Price"] = groupedGamingDF["Average Purchase Price"].astype(float).map("${:,.2f}".format)
groupedGamingDF["Total Purchase Value"] = groupedGamingDF["Total Purchase Value"].astype(float).map("${:,.2f}".format)
groupedGamingDF["Avg Total Purchase per Person"] = groupedGamingDF["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

groupedGamingDF
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
topSpendersDF = pd.DataFrame(gamingDF.groupby("SN")["SN"].count())
topSpendersDF = topSpendersDF.rename(columns={"SN" : "Purchase Count"})
topSpendersDF["Average Purchase Price"] = gamingDF.groupby("SN")["Price"].mean()
topSpendersDF["Total Purchase Value"] = gamingDF.groupby("SN")["Price"].sum()


topSpendersDF = topSpendersDF.sort_values(by=["Purchase Count"], ascending=False)
topSpendersDF["Average Purchase Price"] = topSpendersDF["Average Purchase Price"].astype(float).map("${:,.2f}".format)
topSpendersDF["Total Purchase Value"] = topSpendersDF["Total Purchase Value"].astype(float).map("${:,.2f}".format)

topSpendersDF
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popItemsDF = pd.DataFrame(gamingDF.groupby("Item ID")["Item ID"].count())
popItemsDF = popItemsDF.rename(columns={"Item ID" : "Purchase Count"})

# fixing what price and item name columns look like?
# setting item name as index once that is fixed

popItemsDF["Item Name"] = gamingDF.groupby("Item ID")["Item Name"]
popItemsDF["Price"] = gamingDF.groupby("Item ID")["Price"]
popItemsDF = popItemsDF.rename(columns={"Price" : "Item Price"})
popItemsDF["Total Purchase Value"] = gamingDF.groupby("Item ID")["Price"].sum()

popItemsDF = popItemsDF.sort_values(by=["Purchase Count"], ascending=False)

popItemsDF["Total Purchase Value"] = popItemsDF["Total Purchase Value"].astype(float).map("${:,.2f}".format)
#popItemsDF["Item Price"] = popItemsDF["Item Price"].astype(float).map("${:,.2f}".format)

popItemsDF.head()
```
